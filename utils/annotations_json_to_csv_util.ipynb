```python
import pandas as pd
import numpy as np
```

Read Json file exported from Labelbox

```python
#data=pd.read_json('export-2019-04-02T07_48_47.797Z.json')
data=pd.read_json('export-2019-04-03T16_46_34.116Z_247_imgs_scaled.json')
```

```python
data['Label'][1]
```

Parsing Json and saving bounding box coordinates in below format : <br>
img_path,x_min,y_min,x_max,y_max

```python
data_1 = pd.DataFrame()
data_1['format']= 'YE358311_Fender_apron\\train_fastrcnn\\'+data['External ID']
for i in range(data.shape[0]):
    try:
        x1=data['Label'][i]['Defected'][0]['geometry'][0]['x']
        x2=data['Label'][i]['Defected'][0]['geometry'][2]['x']
        x_min=min(x1,x2)
        x_max=max(x1,x2)
        y1=data['Label'][i]['Defected'][0]['geometry'][0]['y']
        y2=data['Label'][i]['Defected'][0]['geometry'][2]['y']
        y_min=min(y1,y2)
        y_max=max(y1,y2)
        data_1['format'][i] = data_1['format'][i]+ ',' +str(x_min)+','+str(y_min)+','+str(x_max)+','+str(y_max)+','+'defected'
    except:
        data_1['format'][i] = data_1['format'][i]
```
