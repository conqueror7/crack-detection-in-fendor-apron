```python
import cv2
import numpy as np
import pandas as pd
import os
```

```python
def rescale_img(img_path,out_dir):    
    for idx, img_name in enumerate(sorted(os.listdir(img_path))):
        filepath = os.path.join(img_path,img_name)
        img = cv2.imread(filepath,cv2.IMREAD_UNCHANGED) 
        print('Original Dimensions : ',img.shape)
        width = 640
        height = 480
        dim = (width, height)
        # resize image
        resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
        print('Resized Dimensions : ',resized.shape)
        cv2.imwrite(out_dir+'\\{}.jpg'.format(img_name),img)
        #cv2.waitKey(0)
        #cv2.destroyAllWindows() 
```

```python
rescale_img('train\\Defected','train-scaled')
```

```python
rescale_img('train\\Healthy','train-scaled\\Healthy')
```

```python
rescale_img('validation\\Defected','validation-scaled\\Defected')
rescale_img('validation\\Healthy','validation-scaled\\Healthy')
```
